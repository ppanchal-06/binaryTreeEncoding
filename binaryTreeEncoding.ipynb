```python
inputString = "0011000000011000"
# inputString = "0000000000000010"

def binaryTreeEncoder(inputString):
  encodedStr =""
  decoding_key = " "
  if(len(inputString) <=1):
    return ""
  #divide in half first
  middle = len(inputString)//2
  #check if either side has a 1
  leftSide = 0
  rightSide = 0
  for i in range(0,middle):
    if(inputString[i] == '1'):
      leftSide = 1

  for i in range (middle,len(inputString)):
    if(inputString[i]== '1'):
      rightSide =1
#recursively check each section
  if (leftSide and rightSide):
    if len(inputString) <=2:
      encodedStr+='11'
    encodedStr+=binaryTreeEncoder(inputString[0:middle])
    encodedStr+=binaryTreeEncoder(inputString[middle:len(inputString)])

  elif (leftSide):
    if len(inputString) <=2:
      encodedStr+='10'
    encodedStr+=binaryTreeEncoder(inputString[0:middle])
  elif (rightSide):
    if (len(inputString) <=2):
      encodedStr+= "01"
    encodedStr+=binaryTreeEncoder(inputString[middle:len(inputString)])
  else:
    encodedStr+='00'

#bit replacement
  finalencodedStr = ""
  i = 0;
  while i <(len(encodedStr)):
    if (i <len(encodedStr)-3 and encodedStr[i] == '1' and encodedStr[i+1] =='0' and encodedStr[i+2] == '0' and encodedStr[i+3] == '1' ):
      finalencodedStr+='0'
      i+=4
    elif(i <len(encodedStr)-3 and encodedStr[i] == '0' and encodedStr[i+1] =='1' and encodedStr[i+2] == '0' and encodedStr[i+3] == '1'):
      finalencodedStr+='0'
      i+=3
    else:
      finalencodedStr+=encodedStr[i]
      i+=1

  return finalencodedStr

def decode_key(inputString):
  if(len(inputString) ==1):
    return ""
  middle = len(inputString)//2
  leftSide = 0
  rightSide= 0

  for i in range(0,middle):
    if(inputString[i] == '1'):
      leftSide = 1

  for i in range (middle,len(inputString)):
    if(inputString[i]== '1'):
      rightSide =1


  if (leftSide and rightSide):
    return "11" + decode_key(inputString[:middle]) + decode_key(inputString[middle:])
  if (leftSide):
    return "10" + decode_key(inputString[:middle])
  if (rightSide):
    return "01" + decode_key(inputString[middle:])

print(binaryTreeEncoder(inputString),decode_key(inputString))

```

```python
def binaryTreeDecoder(encodedStr,begin=0,end=16,index=0,decodedStr=None):
  if decodedStr is None:
    decodedStr = "0" * end
  if end ==1:
    decodedStr = decodedStr[:begin] + '1' + decodedStr[begin +1:]
    return decodedStr,index

  middle = end // 2

  if encodedStr[index:index+2]=="11":
    index+=2
    decodedStr, index = binaryTreeDecoder(encodedStr, begin, middle, index, decodedStr)
    decodedStr,index = binaryTreeDecoder(encodedStr, begin+ middle, middle, index,decodedStr)

  elif encodedStr[index:index+2] == "10":
    index+=2
    decodedStr, index= binaryTreeDecoder(encodedStr, begin, middle, index,decodedStr)
  elif encodedStr[index:index+2] == '01':
    index+=2
    decodedStr, index= binaryTreeDecoder(encodedStr, begin+middle, middle, index,decodedStr)


  return decodedStr,index
```
